## 15. Write a Pandas program to filter the DataFrame rows to only show carat weight at least 0.3.

#### Filtering rows in Pandas based on specific conditions is a powerful way to select relevant data from your DataFrames. Here are several methods you can use:

**1. Boolean Indexing:**

- This is the most common and efficient method for filtering rows based on simple conditions.
- Create a Boolean Series where `True` indicates rows that meet your criteria.
- Use the Series as an index to select rows from the DataFrame.

```python
import pandas as pd

# Sample DataFrame
data = pd.DataFrame({'age': [25, 30, 22, 40],
                   'city': ['New York', 'London', 'Paris', 'Tokyo'],
                   'salary': [80000, 95000, 72000, 120000]})

# Filter rows where age is greater than 30 and city is 'New York' or 'London'
filtered_df = data[(data['age'] > 30) & (data['city'].isin(['New York', 'London']))]
print(filtered_df)
```

**2. Query Method:**

- Offers a more readable syntax for expressing filtering conditions.
- Built-in functions like `eq`, `ne`, `gt`, `lt` can be used for comparisons.

```python
filtered_df = data.query("age > 30 and (city == 'New York' or city == 'London')")
print(filtered_df)
```

**3. `loc` and `iloc` Accessors:**

- These methods provide label-based (loc) or integer-based (iloc) indexing for selecting rows.
- Can be used with boolean expressions or conditions within square brackets.

```python
# Filter rows where salary is between 80000 and 100000 (inclusive)
filtered_df = data.loc[data['salary'].between(80000, 100000)]
print(filtered_df)

# Filter rows at indices 1 and 3
filtered_df = data.iloc[[1, 3]]
print(filtered_df)
```

**4. `isin` Method:**

- Checks if values in a column are present in a specific list or set.
- Useful for filtering based on multiple categories or values.

```python
filtered_df = data[data['city'].isin(['Paris', 'Tokyo'])]
print(filtered_df)
```

**5. `drop` Method:**

- Can be used with the `inplace` argument to permanently remove rows that don't meet your criteria.

```python
data.drop(data[data['age'] <= 25].index, inplace=True)
print(data)
```

**Tips:**

- Combine these methods to create complex filtering expressions.
- Use parentheses for clarity and operator precedence.
- Consider performance implications for large DataFrames.

Remember to adjust the conditions and data based on your specific needs. I hope this explanation helps you effectively filter rows in Pandas!

````python
diamonds[diamonds['carat']>=0.3]
````

## 19. Write a Pandas program to find the diamonds that are with a Fair or Good or Premium.

#### While the expression `diamonds[(diamonds['cut'] == 'Fair') | (diamonds['cut'] == 'Good') | (diamonds['cut'] == 'Premium')]` would also work to filter diamonds with the desired cuts, using `isin()` is generally preferred in this case due to several reasons:

**1. Readability and Maintainability:**

- `isin()` offers a more concise and readable way to express the condition, especially when dealing with multiple values.
- It's easier to understand the logic and update the list of cuts if needed.

**2. Performance:**

- `isin()` is often more efficient for filtering based on multiple values compared to repeated comparisons with `|`.
- This becomes more significant with larger datasets.

**3. Flexibility:**

- `isin()` allows you to use lists, sets, or even other Series as the reference for comparison.
- This provides more flexibility in filtering based on different criteria or conditions.

**4. Error Prevention:**

- Using `|` requires careful handling of parentheses and operator precedence to ensure correct evaluation, especially when combining multiple conditions.
- `isin()` avoids these potential errors and simplifies the expression.

**However, there are situations where `|` might be useful:**

- When you have a dynamic list of values to compare against.
- For complex logical expressions involving multiple conditions beyond simple membership checks.

In the specific case of filtering diamonds by cut, `isin()` offers a clear, efficient, and maintainable solution. Remember to choose the method that best suits your specific needs and data context.

````python
diamonds[diamonds['cut'].isin(['Fair', 'Good', 'Premium'])]
````
